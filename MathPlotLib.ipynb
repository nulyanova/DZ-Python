```python
import random
import datetime
# import prettytable                  #   
import matplotlib.pyplot as plt     #   

def BubbleSort(A):
    for i in range(len(A)):
        for j in range(len(A)-1-i):
            if A[j] > A[j+1]:
                A[j], A[j+1] = A[j+1], A[j]

def QuickSort(A, fst, lst):
    if fst >= lst:
        return
    pivot = A[fst]
    first_bigger = fst+1
    while first_bigger <= lst:
        if A[first_bigger] >= pivot:
            break
        first_bigger += 1
    i = first_bigger+1
    while i <= lst:
        if A[i] < pivot:
            A[i], A[first_bigger] = A[first_bigger], A[i]
            first_bigger += 1
        i += 1
    last_smaller = first_bigger-1
    A[fst], A[last_smaller] = A[last_smaller], A[fst]
    QuickSort(A, fst, last_smaller-1)
    QuickSort(A, first_bigger, lst)

def InsertionSort(A):
    for i in range(1, len(A)):
        key = A[i]
        j = i - 1
        while j >= 0 and key < A[j]:
            A[j + 1] = A[j]
            j -= 1
        A[j + 1] = key

def ShakerSort(A):
    n = len(A)
    for i in range(n // 2):
        for j in range(i, n - 1 - i):
            if A[j] > A[j + 1]:
                A[j], A[j + 1] = A[j + 1], A[j]
        for j in range(n - 2 - i, i, -1):
            if A[j] < A[j - 1]:
                A[j], A[j - 1] = A[j - 1], A[j]

x = []
y_bubble = []
y_quick = []
y_insert = []
y_shaker = []

for N in range(1000, 5001, 1000):
    x.append(N)
    min = 1
    max = N
    A = [random.randint(min, max) for _ in range(N)]

    A_copy = A.copy()
    B_copy = A.copy()
    C_copy = A.copy()
    D_copy = A.copy()

    t1 = datetime.datetime.now()
    BubbleSort(A_copy)
    t2 = datetime.datetime.now()
    y_bubble.append((t2 - t1).total_seconds())
    print("    " + str(N) + "      " + str((t2 - t1).total_seconds()) + "c")

    t3 = datetime.datetime.now()
    QuickSort(B_copy, 0, len(B_copy) - 1)
    t4 = datetime.datetime.now()
    y_quick.append((t4 - t3).total_seconds())
    print("   " + str(N) + "      " + str((t4 - t3).total_seconds()) + "c")

    t5 = datetime.datetime.now()
    InsertionSort(C_copy)
    t6 = datetime.datetime.now()
    y_insert.append((t6 - t5).total_seconds())
    print("   " + str(N) + "      " + str((t6 - t5).total_seconds()) + "c")

    t7 = datetime.datetime.now()
    ShakerSort(D_copy)
    t8 = datetime.datetime.now()
    y_shaker.append((t8 - t7).total_seconds())
    print("   " + str(N) + "      " + str((t8 - t7).total_seconds()) + "c")

#  
plt.figure(figsize=(10, 6))
plt.plot(x, y_bubble, "C0", label="Bubble Sort")
plt.plot(x, y_quick, "C1", label="Quick Sort")
plt.plot(x, y_insert, "C2", label="Insertion Sort")
plt.plot(x, y_shaker, "C3", label="Shaker Sort")
plt.title("   ")
plt.xlabel(" ")
plt.ylabel(" ()")
plt.legend()
plt.grid(True)
plt.show()

# BarPlot  
plt.figure(figsize=(10, 6))
width = 200
plt.bar([p - width for p in x], y_bubble, width=width, label="Bubble Sort")
plt.bar(x, y_quick, width=width, label="Quick Sort")
plt.bar([p + width for p in x], y_insert, width=width, label="Insertion Sort")
plt.bar([p + 2 * width for p in x], y_shaker, width=width, label="Shaker Sort")
plt.title("    (BarPlot)")
plt.xlabel(" ")
plt.ylabel(" ()")
plt.legend()
plt.grid(True)
plt.show()

#     
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

axs[0, 0].plot(x, y_bubble, "C0")
axs[0, 0].set_title("Bubble Sort")
axs[0, 0].set_xlabel(" ")
axs[0, 0].set_ylabel(" ()")

axs[0, 1].plot(x, y_quick, "C1")
axs[0, 1].set_title("Quick Sort")
axs[0, 1].set_xlabel(" ")
axs[0, 1].set_ylabel(" ()")

axs[1, 0].plot(x, y_insert, "C2")
axs[1, 0].set_title("Insertion Sort")
axs[1, 0].set_xlabel(" ")
axs[1, 0].set_ylabel(" ()")

axs[1, 1].plot(x, y_shaker, "C3")
axs[1, 1].set_title("Shaker Sort")
axs[1, 1].set_xlabel(" ")
axs[1, 1].set_ylabel(" ()")

plt.tight_layout()
plt.show()
```
